# Matplotlib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK. There is also a procedural "pylab" interface based on a state machine (like OpenGL), designed to closely resemble that of MATLAB, though its use is discouraged. SciPy makes use of Matplotlib.

Matplotlib was originally written by John D. Hunter. Since then it has had an active development community and is distributed under a BSD-style license. Michael Droettboom was nominated as matplotlib's lead developer shortly before John Hunter's death in August 2012 and was further joined by Thomas Caswell. Matplotlib is a NumFOCUS fiscally sponsored project.

One can install matplotlib via `pip`:

```
pip install matplotlib
```

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias.

````python
import numpy as np
import matplotlib.pyplot as plt
````

## Line Plot

The `plot()` function is used to draw points (markers) in a diagram. By default, the `plot()` function draws a line from point to point. The function takes parameters for specifying points in the diagram. Parameter 1 is an array containing the points on the x-axis. Parameter 2 is an array containing the points on the y-axis. You can plot as many points as you like, just make sure you have the same number of points in both axis. Of course, you can plot as many lines as you want by including multiple `plot()` functions. In order for the plot to be plotted you need to use the `plt.show()` function.

Each line plot has a huge list of congifuration one can change, that makes the lines look different. It is impossible to go through all of them, however I'll go through some of the most common ones.

* The `color` value, specifies the color of the line. It can be one of the basic colours, or we can also use Hexadecimal color values, or any of the 140 supported color names. Here is a list with the basic colors. (For a full list of possible colors check [here](https://matplotlib.org/stable/gallery/color/named_colors.html))
    * `r`: Red
    * `g`: Green
    * `b`: Blue
    * `c`: Cyan
    * `m`: Magenta
    * `y`: Yellow
    * `k`: Black
    * `w`: White

* The keyword argument `linestyle` defines the style of the line. Here is a full list of possible linestyles.
    * `-`: Solid Line
    * `:`: Dotted Line
    * `--`: Dashed Line
    * `-.`: Dashed/dotted Line

* The keyword argument `linewidth` to set the width of the line. 
* You can adjust the transparency of the lines with the `alpha` argument. 
* You can use the keyword argument `marker` to emphasize each point with a specified marker. For a full list of possible markers check [here](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers).

On top of plotting, matplotlib offers also a lot of option for basic plot configuration. Once again, the list is endless since every possible apsect of a plot can be manipulated. Here I will go through some of the most usefull ones.

* First of all, one can use the function `plt.figure()` on top of the plot to specify characteristics like the size of the plot and many more. Here we will define only the size of the figure with the keyword `figsize`.

* With Pyplot, you can use the `title()` function to set a title for the plot, and the `xlabel()` and `ylabel()` functions to set a label for the x-axis and y-axis. On top of that you can use the `loc` parameter in all of these functions to position the names. Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

* When plotting multiple plots, each plot can carry a unique `label` that will make it easy to identify. With the keyword `plt.legend()` you can plot these labels. The plots, by default, are plotted with different colors.

* With Pyplot you can use `xlim()` and `ylim()` functions to specify the limits of x-axis and y-axis. You can also use `xticks()` and `yticks()` functions to specify the ticks you want to appear in the x and y axis.

````python
# create data to plot
x = np.linspace(1,20, num=20)
y1 = np.sort(np.random.random_sample(size=20) * 100)
y2 = np.sort(np.random.random_sample(size=20) * 100 + 50)
````

````python
plt.figure(figsize=(14,4))
plt.plot(x, y1, label="y1", color = 'r', alpha=1, marker="X", linestyle = '--', linewidth = 1.5)
plt.plot(x, y2, label="y2", color = 'b', alpha=0.5, marker=".", linestyle = '-', linewidth = 3)
plt.legend()
plt.title("A Title", loc = 'left')
plt.xlabel("X")
plt.xlim([0,21])
plt.xticks([i for i in range(20)])
plt.ylabel("Y") 
plt.ylim([0,150])
plt.show()
````

## Scatter Plot
With Pyplot, you can use the `scatter() `function to draw a scatter plot.

The `scatter()` function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis. As before you can add multiple scatter plots in the same plot by using `scatter()` function multiple times. The plots, by default, are plotted with different colors.

All available options from line plots can also be found in scatter plots. For example, you can set your own color for each scatter plot with the `color` argument. You can change the size of the dots with the `size` argument. You can adjust the transparency of the dots with the `alpha` argument.

````python
# create data to plot
x = np.linspace(1,20, num=20)
y1 = np.sort(np.random.random_sample(size=20) * 100)
y2 = np.sort(np.random.random_sample(size=20) * 100 + 50)

# plot
plt.scatter(x, y1, label="y1", alpha=1, marker="X",)
plt.scatter(x, y2, label="y2", alpha=1, marker=".")
plt.show()
````

## Bar Plot

With Pyplot, you can use the `bar()` function to draw bar graphs. The `bar()` function takes arguments that describes the layout of the bars. The categories and their values represented by the first and second argument as arrays. If you want the bars to be displayed horizontally instead of vertically, use the `barh()` function. Both `bar()` and `barh()` takes the keyword argument color to set the color of the bars. You can use any of the 140 supported color names. The `bar()` takes the keyword argument width to set the width of the bars (the default width value is 0.8). For horizontal bars, use height instead of width.

````python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color = "red", width = 0.1)
plt.show()
````

## Histogram

A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval. In Matplotlib, we use the `hist()` function to create histograms. The `hist()` function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

````python
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
````

## Pie Chart

With Pyplot, you can use the `pie()` function to draw pie charts. The pie chart draws one piece (called a wedge) for each value in the array. By default the plotting of the first wedge starts from the x-axis and move counterclockwise. The size of each wedge is determined by comparing the value with all the other values, by using the value divided by the sum of all values.

We can add labels to the pie chart with the label parameter. The label parameter must be an array with one label for each wedge.

As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter. The startangle parameter is defined with an angle in degrees, default angle is 0.

Maybe you want one of the wedges to stand out? The explode parameter allows you to do that. The explode parameter, if specified, and not None, must be an array with one value for each wedge. Each value represents how far from the center each wedge is displayed.

Add a shadow to the pie chart by setting the shadows parameter to True.

You can set the color of each wedge with the colors parameter. The colors parameter, if specified, must be an array with one value for each wedge.

To add a list of explanation for each wedge, use the `legend()` function.

````python
y = np.array([35, 25, 25, 15])

plt.pie(y, 
        labels = ["Apples", "Bananas", "Cherries", "Dates"], 
        startangle = 90, 
        explode = [0.2, 0, 0, 0], 
        shadow = True, 
        colors = ["black", "hotpink", "b", "#4CAF50"])
plt.legend()
plt.show() 
````

## Subplots

With the `subplot()` function you can draw multiple plots in one figure. The `subplot()` function takes three arguments that describes the layout of the figure. The layout is organized in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot.

For example, `plt.subplot(1, 2, 1)` means that the figure has 1 row, 2 columns, and this plot is the first plot. `plt.subplot(1, 2, 2)` means that the figure has 1 row, 2 columns, and this plot is the second plot.

So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can use those 2 functions.

You can draw as many plots you like on one figure, just descibe the number of rows, columns, and the index of the plot.

You can add a title to the entire figure with the `suptitle()` function

You can add a configuration to each plot with the configuration functions we introduced previously.

````python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.suptitle("Generic Title")
plt.show()
````
